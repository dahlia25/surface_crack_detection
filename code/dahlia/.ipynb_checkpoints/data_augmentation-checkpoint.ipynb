# Surface Crack Images - Data Augmentation
This notebook contains the code to augment and save image data for training deep learning models

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.preprocessing.image import load_img, save_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

no need to resize images since data image size is uniform

```python
# checking to see if listing the correct directory with the images
img_files = sorted(os.listdir(r'data/archive/original/0'))
img_files[:5]
```

```python
## exploring ImageDataGenerator function

img_folder = 'data/archive/original/1/'
img_file = img_files[0]

img = load_img(img_folder+img_file)

data = img_to_array(img)

samples = np.expand_dims(data, 0)

datagen = ImageDataGenerator(horizontal_flip=True, 
                             zoom_range=[0.5,3], 
                             brightness_range=[0.2, 0.6], 
                             rotation_range=90,
                             width_shift_range=0.5,
                             height_shift_range=0.5,
                             fill_mode='reflect')

it = datagen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330+1+i)
    batch=it.next()
    
    image = batch[0].astype('uint8')
    
    plt.imshow(image)
    
plt.show()
```

Augment positive image data (has crack on surface)

```python
## automated data augmentation generation with image saving to folder

# create data augmentation generator
datagen = ImageDataGenerator(horizontal_flip=True, 
                             zoom_range=[0.5,3], 
                             brightness_range=[0.2, 0.6], 
                             rotation_range=90,
                             width_shift_range=0.5,
                             height_shift_range=0.5,
                             fill_mode='reflect')


# generate augmented data
pos_img_save_folder = 'data/archive/augmented/1/'
pos_img_src_folder = 'data/archive/original/1/'
pos_img_files = sorted(os.listdir(pos_img_src_folder))

counter = 1

for img_file in pos_img_files:

    img = load_img(pos_img_src_folder+img_file)
    data = img_to_array(img)
    samples = np.expand_dims(data, 0)
    it = datagen.flow(samples, batch_size=1)

    for i in range(9):
        batch=it.next()   
        image = batch[0].astype('uint8')
        
        ## for img file naming convention
        lead_zeros = '00000'
        zeros_to_remove = len(str(counter)) - 1
        zeros_index = len(lead_zeros) - zeros_to_remove
        img_num = lead_zeros[0:zeros_index] + str(counter)
        
        save_img(f'{pos_img_save_folder}{str(img_num)}.jpg', image)
        
        if counter % 1000 == 0:
            print(f'{str(img_num)}.jpg')
        
        counter += 1

check = sorted(os.listdir(pos_img_save_folder))
print(f'{len(check)} images augmentation completed..')
```

Augment negative image data (has no surface crack)

```python
## automated data augmentation generation with image saving to folder

# create data augmentation generator
datagen = ImageDataGenerator(horizontal_flip=True, 
                             zoom_range=[0.5,3], 
                             brightness_range=[0.2, 0.6], 
                             rotation_range=90,
                             width_shift_range=0.5,
                             height_shift_range=0.5,
                             fill_mode='reflect')


# generate augmented data
neg_img_save_folder = 'data/archive/augmented/0/'
neg_img_src_folder = 'data/archive/original/0/'
neg_img_files = sorted(os.listdir(neg_img_src_folder))

counter = 1

for img_file in neg_img_files:

    img = load_img(neg_img_src_folder+img_file)
    data = img_to_array(img)
    samples = np.expand_dims(data, 0)
    it = datagen.flow(samples, batch_size=1)

    for i in range(9):
        batch=it.next()   
        image = batch[0].astype('uint8')
        
        ## for img file naming convention
        lead_zeros = '00000'
        zeros_to_remove = len(str(counter)) - 1
        zeros_index = len(lead_zeros) - zeros_to_remove
        img_num = lead_zeros[0:zeros_index] + str(counter)
        
        save_img(f'{neg_img_save_folder}{str(img_num)}.jpg', image)
        
        if counter % 1000 == 0:
            print(f'{str(img_num)}.jpg')
        
        counter += 1

check = sorted(os.listdir(neg_img_save_folder))
print(f'{len(check)} images augmentation completed..')
```

**Check number of images in original and augmented folders**

```python
total_neg_img = len(os.listdir(neg_img_src_folder))
total_pos_img = len(os.listdir(pos_img_src_folder))
total_aug_neg_img = len(os.listdir(neg_img_save_folder))
total_aug_pos_img = len(os.listdir(pos_img_save_folder))

print(f'Total negative images = neg {total_neg_img} + augmented neg {total_aug_neg_img} = {total_neg_img + total_aug_neg_img}')
print(f'Total positive images = pos {total_pos_img} + augmented neg {total_aug_pos_img} = {total_pos_img + total_aug_pos_img}')
```
