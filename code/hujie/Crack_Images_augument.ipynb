```python
def image_augmenter():
    return ImageDataGenerator(
        horizontal_flip=True,
        brightness_range=[0.2, 0.6]
        )
```

```python
def augmentation_preview(actual_img, samples, f):
    plt.subplot(331)
    img_to_array(actual_img)
    pyplot.imshow(actual_img)
    it = image_augmenter().flow(samples, batch_size=1)
    for i in range(9):
        pyplot.subplot(330 + 1 + i)
        plt.subplots_adjust(left=0.0, right=1.0, bottom=0.0, top=1.0)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
    pyplot.savefig(os.path.join("previews","prev_" + f))
    plt.show()
```

```python
def active_augmentation(path, filenme, samples):
    it = image_augmenter().flow(samples, batch_size=1)
    for i in range(5):
        batch = it.next()
        image = batch[0].astype('uint8')
        target_filename = os.path.join(path, "Aug_" + str(i) + "_" + filenme)
        save_img(target_filename, image)
```

```python
def main(input_dir, output_dir):
    print("Input dir  : {}".format(input_dir))
    print("Output dir : {}".format(output_dir))
    for f in os.listdir(input_dir):
        img = load_img(os.path.join(input_dir, f))
        samples = np.expand_dims(img_to_array(img), 0)
        print(samples.shape)
        # do_augmentation_preview(img, samples, f)
        active_augmentation(output_dir, f, samples)
```
